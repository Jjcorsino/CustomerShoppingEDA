```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('shopping_trends.csv', encoding='latin1')
df.head()
colors = ["#89CFF0", "#FF69B4", "#FFD700", "#7B68EE", "#FF4500",
          "#9370DB", "#32CD32", "#8A2BE2", "#FF6347", "#20B2AA",
          "#FF69B4", "#00CED1", "#FF7F50", "#7FFF00", "#DA70D6"]
```

```python
df.isnull().sum()
```

```python
sns.countplot(
    x='Gender',
    data=df,
    palette='hls',
    hue= 'Gender',
    dodge='Auto',
)
```

```python
plt.figure(figsize=(20, 6))

counts = df['Gender'].value_counts()
explode = (0, 0.1)

plt.pie(
    counts,
    labels=counts.index,
    explode=explode,
    autopct='%1.1f%%',
    startangle=90,
)
```

```python
plt.figure(figsize=(20, 6))

counts = df['Category'].value_counts()

plt.pie(
    counts,
    labels=counts.index,
    autopct='%1.1f%%',
    startangle=90,
)
```

```python
df['Item Purchased'].value_counts()
```

```python
plt.figure(figsize=(20, 6))

counts = df['Item Purchased'].value_counts().nlargest(10)

plt.bar(
    counts.index,
    counts,
    color=colors,
)
```

```python
plt.figure(figsize=(20, 6))

counts = df['Location'].value_counts().nlargest(10)

plt.bar(
    counts.index,
    counts,
    color=colors,
)
```
